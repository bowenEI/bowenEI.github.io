# pandas 运算

```python
import pandas as pd
import numpy as np
```

## 算术运算

```python
df = pd.DataFrame(np.random.randn(10, 4), columns=["A", "B", "C", "D"])
df
```

```python
df = pd.DataFrame(np.random.randn(7, 3), columns=["A", "B", "C"])
df
```

```python
df * 5 + 2
```

```python
1 / df
```

```python
df ** 4
```

## 布尔运算

```python
df = pd.DataFrame({"a": [1, 0, 1], "b": [0, 1, 1]}, dtype=bool)
df
```

```python
df = pd.DataFrame({"a": [0, 1, 1], "b": [1, 1, 0]}, dtype=bool)
df
```

```python
df & df
```

```python
df | df
```

```python
df ^ df
```

```python
-df
```
